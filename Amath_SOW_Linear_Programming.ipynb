# Linear Programming using CVXPY

```python
# Make sure you have the package cvxpy installed
from cvxpy import *
import numpy as np
```

# Exercise 1

A carpenter makes tables and chairs. Each table can be sold for a profit of £30 and each chair for a profit of £10. The carpenter can afford to spend up to 40 hours per week working and takes six hours to make a table and three hours to make a chair. Customer demand requires that he makes at least three times as many chairs as tables. Tables take up four times as much storage space as chairs and there is room for at most four tables each week. 

```python
# Put your solution here
x = Variable() #nomber of table
y = Variable() # nomber of chairs
z=30*x + 10*y
objective = Maximize(z)
constraints = [6*x+3*y<=40,y>=3*x,x+0.25*y<=4,x>=0,y>=0]

reslt=Problem(objective, constraints)
result=reslt.solve()

print('Status:' , reslt.status)
print('Maximum anomber of chairs and table made by week:' , np.round(reslt.value))
print('number of table:' , np.round(x.value))
print('number of chairs:' , np.round(y.value))
```

# Exercise 2

```python
# Put your solution here
x = Variable() #produced by A
y = Variable() # produced by B
z=3*x + 5*y
objective = Maximize(z)
constraints = [12*x+25*y<=1800,y>=0.4*x,x>=0,y>=0]

reslt=Problem(objective, constraints)
result=reslt.solve()

print('Status:' , reslt.status)
print('Maximum amount to pay per week:' , np.round(reslt.value))
print('payment for production A:' , np.round(x.value))
print('payment for production B:' , np.round(y.value))
```
